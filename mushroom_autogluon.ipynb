```python
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train_df = pd.read_csv('/kaggle/input/playground-series-s4e8/train.csv').drop('id', axis=1)
orig_df = pd.read_csv("/kaggle/input/secondary-mushroom-dataset-data-set/MushroomDataset/secondary_data.csv", sep=";")
train_df = pd.concat([train_df, orig_df], ignore_index=True)
train_df.head()
```

```python
test_df = pd.read_csv('/kaggle/input/playground-series-s4e8/test.csv').drop('id', axis=1)
test_df.head()
```

```python
predictor = TabularPredictor(
    label='class',
    eval_metric='mcc',
    problem_type='binary'
).fit(
    train_df,
    presets='best_quality',
    time_limit=3600*10,
    ag_args_fit={'num_gpus': 2},  # 두 개의 GPU와 16개의 CPU 사용
    auto_stack=True,
    refit_full=True,
    verbosity=3)
results = predictor.fit_summary()
```

```python
predictor.leaderboard()
```

```python
y_pred = predictor.predict(test_df)
```

```python
sub = pd.read_csv('/kaggle/input/playground-series-s4e8/sample_submission.csv')
sub['class'] = y_pred
sub.to_csv('submission.csv', index=False)
```
